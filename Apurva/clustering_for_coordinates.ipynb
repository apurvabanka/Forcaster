```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../food_delivery_dataset/train.csv')

df = df.replace('NaN', np.nan, regex=True)

df['Restaurant_latitude'] = pd.to_numeric(df['Restaurant_latitude'])
df['Restaurant_longitude'] = pd.to_numeric(df['Restaurant_longitude'])
df['Delivery_location_latitude'] = pd.to_numeric(df['Delivery_location_latitude'])
df['Delivery_location_longitude'] = pd.to_numeric(df['Delivery_location_longitude'])

df = df[~((df['Restaurant_latitude'] <= 1) & (df['Restaurant_longitude'] <= 1) &(df['Delivery_location_latitude'] <= 1) &(df['Delivery_location_longitude'] <= 1))]

df = df.head(100)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Restaurant_latitude'], df['Restaurant_longitude'], color='red', marker='o')
plt.scatter(df['Delivery_location_latitude'], df['Delivery_location_longitude'], color='blue', marker='o', alpha=0.2)

# Adding labels
plt.title('Latitude and Longitude Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Display the plot
plt.grid(True)
plt.show()
```
